```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import datetime
```

```python
rootPath = "DailyData"
dirlist = []
for fd in os.listdir(rootPath): 
    if os.path.isdir(os.path.join(rootPath,fd)): # not .zip or .DS_Store
        dirlist.append(fd) # 201301
dirlist.sort(key= lambda x:int(x))  #sort by date for MacOS

allCsv = pd.DataFrame(columns=['date','PM2.5', 'PM10', 'SO2', 'NO2',
           'CO', 'O3', 'U(m/s)', 'V(m/s)', 'TEMP(K)',
           'RH(%)', 'PSFC(Pa)','lat','lon'])
for ds in dirlist:
#     print(ds) # 201301 dir
    files = os.listdir(os.path.join(rootPath,ds)) # CN-Reanalysis-daily-2013010100.csv
    files.sort(key= lambda x:int(x[25:28]))  #sort by date for MacOS
  
    for csv in files: 
#         print(csv) # CN-Reanalysis-daily-2013010100.csv
        time = csv.split("-")[-1].split(".")[0]
        date = datetime.date(int(time[:4]), int(time[4:6]), int(time[6:8]))
        print(date)

        csv_data = pd.read_csv(os.path.join(rootPath,ds,csv))
        csv_data = csv_data.drop(columns=[" "])
        csv_data.columns = ['PM2.5', 'PM10', 'SO2', 'NO2',
           'CO', 'O3', 'U(m/s)', 'V(m/s)', 'TEMP(K)',
           'RH(%)', 'PSFC(Pa)','lat','lon']
        csv_data['date'] = date
#         allCsv = allCsv.append(csv_data.loc[0::210],ignore_index=True)
        allCsv = allCsv.append(csv_data.loc[:],ignore_index=True)
    
allCsv.to_csv('Points202.csv',  encoding='utf_8_sig', index=False)
    
    
```

```python
monthlyMeanAll = pd.DataFrame(columns=['date', 'PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'U', 'V', 'TEMP',
       'RH', 'PSFC', 'WindSpeed'])
```

```python
rootpath = 'yearlyData'
for f in os.listdir(rootpath):
    c = pd.read_csv(os.path.join(rootpath,f))
    c.columns = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'U', 'V', 'TEMP',
       'RH', 'PSFC', 'WindSpeed']
    c['date'] = [pd.to_datetime(f[:-4]+str(101+index)[1:]+"01") for index in range(12)]
    monthlyMeanAll = monthlyMeanAll.append(c,ignore_index=True)
    
```

```python
monthlyMeanAll
```

```python
monthlyMeanAll.to_sql('monthlyMeanAll', engine, index= False)
```

```python
monthlyMeanAll_norm = (monthlyMeanAll-monthlyMeanAll.min())/(monthlyMeanAll.max()-monthlyMeanAll.min())
```

```python
monthlyMeanAll_norm
```

```python
monthlyMeanAll_norm['date'] = monthlyMeanAll['date']
```

```python
monthlyMeanAll_norm
```

```python
monthlyMeanAll_norm.to_sql('monthly_Mean_All_norm', engine, index= False)
```
