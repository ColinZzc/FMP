```python
## 数据库加一列 alter table locations add (elevation NUMBER(6,2));
## alter table locations add constraint pk_latlon primary key (lat, lon)
from PIL import Image
from io import BytesIO
import urllib
import math

def deg2num(lat_deg, lon_deg, zoom):
  lat_rad = math.radians(lat_deg)
  n = 2.0 ** zoom
  xtile = int((lon_deg + 180.0) / 360.0 * n)
  ytile = int((1.0 - math.asinh(math.tan(lat_rad)) / math.pi) / 2.0 * n)
  return (xtile, ytile)

def cor2ele(lat, lon):
    zoom = 30
    x, y = deg2num(lat, lon, 30)
    token = 'pk.eyJ1Ijoia2FyYTI3OSIsImEiOiJja3diaXlpMDhnaTVnMm9wZ3h6OTl4ZGtpIn0.PT0tVC9Lmd0wB0NIPBHxLg'
    url = f'''https://api.mapbox.com/v4/mapbox.terrain-rgb/{zoom}/{x}/{y}.pngraw?access_token={token}'''
    with urllib.request.urlopen(url) as d:
        data = d.read()
        img = Image.open(BytesIO(data))
        pix = img.load()
        R,G,B,_ = pix[0,0]
        height = -10000 + ((R * 256 * 256 + G * 256 + B) * 0.1)
    return height

import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy.orm import sessionmaker
from sqlalchemy import create_engine
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column,Float

#创建对象基类
Base = declarative_base()

#location对象
class Location(Base):
    #表名
    __tablename__ = 'locations'
    #表结构
    lat = Column(Float, primary_key=True)
    lon = Column(Float, primary_key=True)
    elevation = Column(Float)

engine = create_engine('oracle+cx_oracle://colin:colin@localhost:1521/?service_name=airvispdb.mshome.net', echo=False)
# session类
DBsession = sessionmaker(bind=engine)
#session对象
session = DBsession()

#跟新直接修改类就好
```

```python
locationList = session.query(Location)
locationList
```

```python
index = 0
for loca in locationList:
    print(loca.lat)
    print(loca.lon)
    print(loca.elevation)
    
    index+=1
    if index>2:
        break
```

```python
for index, loca in enumerate(locationList):
    if index%1000==0:
        session.commit()
    try:
        if loca.elevation is None:
            lat = loca.lat
            lon = loca.lon
            ele = cor2ele(lat, lon)
            ele = round(ele, 2)
            print(index, lat, lon, ele)
            loca.elevation = ele
    except:
        pass
    finally:
        session.commit()
```
