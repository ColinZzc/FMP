### 计算月平均库 线性模板 每月一张表 40s一张 12*6=72张

```python
import numpy as np
import pandas as pd
from sqlalchemy import create_engine


engine = create_engine('oracle+cx_oracle://colin:colin@localhost:1521/?service_name=airvispdb.mshome.net', echo=False)

def process_bar(percent, start_str='', end_str='', total_length=0):
    bar = ''.join(["="] * int(percent * total_length)) + ''
    bar = '\r' + start_str + bar.ljust(total_length) + ' {:0>4.1f}%|'.format(percent*100) + end_str
    print(bar, end='', flush=True)

for year in range(2013,2019): #13-18年
    for month in range(1,12): #1-11月

        sql = f'''
                create table {"avg"+str(year)}{month:02d} as 
                    select avg(pm25) as avg_pm25, 
                            avg(pm10) as avg_pm10, 
                            avg(so2) as avg_so2, 
                            avg(no2) as avg_no2, 
                            avg(co) as avg_co, 
                            avg(o3) as avg_o3,
                            avg(u) as avg_u, 
                            avg(v) as avg_v,
                            avg(temp) as avg_temp,
                            avg(rh) as avg_rh,
                            avg(psfc) as avg_psfc,
                            lat,
                            lon
                    from airvis where day >= Date '{year}-{month:02d}-01' and day < Date '{year}-{month+1:02d}-01' group by lat,lon
                '''
        try:
            pd.read_sql_query(sql, engine)
            print(year, month, "Done")
        except Exception as e:
            print("error", year, month)
            print(e)
        # print(sql)
        # break

    # 12月 sql where 不一样
    sql = f'''
                create table avg{str(year)}12 as 
                    select avg(pm25) as avg_pm25, 
                            avg(pm10) as avg_pm10, 
                            avg(so2) as avg_so2, 
                            avg(no2) as avg_no2, 
                            avg(co) as avg_co, 
                            avg(o3) as avg_o3,
                            avg(u) as avg_u, 
                            avg(v) as avg_v,
                            avg(temp) as avg_temp,
                            avg(rh) as avg_rh,
                            avg(psfc) as avg_psfc,
                            lat,
                            lon
                    from airvis where day >= Date '{year}-12-01' and day <= Date '{year}-12-31' group by lat,lon
                '''

    try:
        pd.read_sql_query(sql, engine)
        print(year, month, "Done")
    except Exception as e:
        print("error", year, month)
        print(e)
 
    # print(sql)
```
