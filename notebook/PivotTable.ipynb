```python
data_result = pd.pivot_table(data, index=["year", "month", "name"], 
                             values=['PM2.5', 'PM10', 'SO2', 'NO2', 
                                      'CO', 'O3', 'TEMP','RH', 'PSFC', 
                                      'lat', 'lon'], 
                             aggfunc={"PM2.5": "mean", "PM10": "mean", 
                                     "SO2": "mean", "NO2": "mean", 
                                     "CO": "mean", "O3": "mean", 
                                     "TEMP": "mean", "RH": "mean", 
                                     "PSFC": "mean",
                                     'lat': "mean", 'lon': "mean"})
data_result
```

```python
# data_result[('2013-01-01','上海')]
# location = ['上海', '北京']
location = list(pd.unique(data['name']))
year = [2018]
filter_result = data_result.query('year ==  '+str(year)+' and name == '+str(location))
filter_result
```

```python
year_result = pd.pivot_table(filter_result.reset_index(), index=["year", "name"], 
                             values=['PM2.5', 'PM10', 'SO2', 'NO2', 
                                      'CO', 'O3', 'TEMP','RH', 'PSFC', 
                                      'lat', 'lon'], 
                             aggfunc={"PM2.5": "mean", "PM10": "mean", 
                                     "SO2": "mean", "NO2": "mean", 
                                     "CO": "mean", "O3": "mean", 
                                     "TEMP": "mean", "RH": "mean", 
                                     "PSFC": "mean",
                                     'lat': "mean", 'lon': "mean"})
year_result
```

```python
res = year_result.reset_index()
res
```

```python
res.columns
```

```python
res[['CO', 'NO2', 'O3', 'PM10', 'PM2.5', 'PSFC', 'RH', 'SO2','TEMP']] = res[['CO', 'NO2', 'O3', 'PM10', 'PM2.5', 'PSFC', 'RH', 'SO2','TEMP']].apply(lambda x: (x - np.min(x)) / (np.max(x) - np.min(x)))
```

```python
res
```

```python
jd = res.to_json(orient="records", force_ascii=False)
jd
```

```python
fh = open('2018CapitalMean.json', 'w', encoding='utf-8')
fh.write(jd)
fh.close()
```

```python
import json
```

```python
fh = open('2018CapitalMean.json', 'r', encoding='utf-8')
jd = fh.readlines()
fh.close()
```

```python
jd = json.loads(jd[0])
```

```python
jd
```

```python
nn_jd = {'name':'地点'}
```

```python
location = []
for i in jd:
    temp = {}
    temp['name'] = i['name']
    temp['lat'] = i['lat']
    temp['log'] = i['lon']
    temp['name'] = i['name']
    temp['group'] = 'A'
    temp['size'] = i['PM2.5']
    location.append(temp.copy())
    temp['group'] = 'B'
    temp['size'] = i['PM10']
    location.append(temp.copy())
    temp['group'] = 'C'
    temp['size'] = i['SO2']
    location.append(temp.copy())
    temp['group'] = 'D'
    temp['size'] = i['NO2']
    location.append(temp.copy())
    temp['group'] = 'E'
    temp['size'] = i['CO']
    location.append(temp.copy())
    temp['group'] = 'F'
    temp['size'] = i['O3']
    location.append(temp.copy())
    
```

```python
location
```

```python
nn_jd['location'] = location
```

```python
nn_jd
```

```python
nn_jd = json.dumps(nn_jd,ensure_ascii=False )
```

```python
fh = open('2018.json', 'w', encoding='utf-8')
jd = fh.write(nn_jd)
fh.close()
```
